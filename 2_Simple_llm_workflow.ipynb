```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
load_dotenv()
```

```python
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
#create a state

class LLMState(TypedDict):
    question:str
    answer:str
```

```python
def llm_res(state:LLMState)->LLMState:
    query=state["question"]
    prompt=f"Answer this question: {query}"
    state["answer"]=model.invoke(prompt).content

    return state
```

```python
# create a graph

graph=StateGraph(LLMState)

#add nodes and edges

graph.add_node("llm_res",llm_res)

graph.add_edge(START,"llm_res")
graph.add_edge("llm_res",END)

workflow=graph.compile()
```

```python
#test
inital_State={"question":"who is president of india"}
workflow.invoke(inital_State)
```
